## Matplotlib

Matplotlib is a plotting library. It helps us turn our data into charts and figures that can be easily understood

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
plt.plot()
```

```python
plt.plot()
plt.show()
```

```python
plt.plot([1, 2, 3, 4]);
```

```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
plt.plot(x, y)
```

### Method One

```python
fig = plt.figure() # creates a figure
ax = fig.add_subplot() # adds some axes
plt.show()
```

### Method Two

```python
fig = plt.figure() # creates a figure
ax = fig.add_axes([1, 1, 1, 1]) # Add axes
ax.plot(x, y) # Add some data
```

### Method Three (Recommended)

```python
fig, ax = plt.subplots()
ax.plot(x, [50, 100, 200, 250]) # add some data
```

###### Matplotlib example workflow

```python
#1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

#2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10))

#3. Plot data
ax.plot(x, y)

#4. Customize plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

#5. Save & show figure
fig.savefig("images/sample1-plot.png")
```

### Making figures with NumPy arrays

##### Bar plot

```python
# Making a plot from a dictionary
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
```

```python
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title="Dan's Nut Butter Store", ylabel='Price($)');
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

##### Histogram

```python
# Make some data
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Options for making Subplots

##### Option One

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2,
                                            figsize=(10, 5))
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2,
                                            figsize=(10, 5))
ax1.plot(x, x/2);
ax1.set(title="Plot1");

ax2.scatter(np.random.random(10), np.random.random(10));
ax2.set(title="Plot2");

ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax3.set(title="Plot3");

ax4.hist(np.random.randn(1000));
ax4.set(title='Plot4');
```

##### Option Two

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot to each different axis/index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```
