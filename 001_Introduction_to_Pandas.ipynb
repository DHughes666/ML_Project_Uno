```python
seat_column = pd.Series([5,5,5,5,5])

# New column called seats
car_sales['Seats'] = seat_column
car_sales
```

```python
car_sales["Seats"].fillna(5, inplace=True)
```

```python
car_sales
```

```python
# Column from Python list
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 4.5, 8.8, 3.9, 3.8, 11]
car_sales["Fuel per 100KM"] = fuel_economy
```

```python
car_sales
```

```python
car_sales["Total fuel used (L)"] = car_sales["Odometer"]/100 * car_sales["Fuel per 100KM"]
```
